```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("shopping_behavior_updated.csv")
data.head()
```

```python
print(f"Data Shape: {data.shape}")
print(f"Data Column Types: \n{data.dtypes}")
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# Previous Purchase and Purchase Amount
df_prev_n_amount = data[["Age", "Purchase Amount (USD)"]]
```

```python
df_prev_n_amount.head()
```

```python
df_sample = df_prev_n_amount.sample(frac=0.1, random_state=42)
x = np.array(df_sample["Age"])
y = np.array(df_sample["Purchase Amount (USD)"])

plt.scatter(x, y)
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.title("Age vs Purchase Amount")

slope, intercept = np.polyfit(x, y, 1)
trend_line = slope * x + intercept
plt.plot(x, trend_line, color='red', label=f"Trend Line (y={slope:.2f}x + {intercept:.2f})")

correlation = df_sample['Age'].corr(df_sample['Purchase Amount (USD)'])
print(f"Correlation Coefficient (Pearson): {correlation:.2f}")
```
